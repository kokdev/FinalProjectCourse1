### 7. Compute the difference between the maximum and minimum elevations for each canton. Find the top 5 cantons that have the largest range of elevations.

Grouped by canton and used aggregate function, the lambda expression to find the range of elevation. Then sorted the value based on the range and got the first 5 cantons. 

It took me time to narrow down this, posted a question in a chapter and the answer helped me to sort on the aggregated values.

Observations:
* Canton of Valais seems to have the largest range of elevations.
* Followed by Tichino Canton

```python
grouped_df  = df.groupby("Canton").agg(
        {
        #"Elevation": lambda x: str(min(x)) + "-" + str(max(x)), 
        "Elevation": [lambda x: str(min(x)) + "-" + str(max(x)), lambda x: max(x) - min(x)]
    }
).sort_values(("Elevation", "<lambda_1>"), ascending=False).head(5)
grouped_df.rename(columns={"<lambda_0>": "Range", "<lambda_1>": "Diff"}).reset_index()

```

### 3. We are interested in the number of communes and their proportions of young residents (0-19yrs). Your task is to obtain a histogram that shows the number of communes for which this proportion falls into the following intervals: 
* 5%-10%
* 10%-15%
* 15%-20%
* 20%-25%
* 25%-30%
* 30%-35%

Calculated the data for the 0-19 years residents using the appropriate percentages and then used them to pot the histogram. Used the figsize to set the width and height and bin for the setting the boundaries of the plot.

Observation:
* 20-25% of the 0-19 years young residents has the maximum number of communes.

```python
df = pd.read_csv("p1_communes.csv")
df = df.reset_index().set_index("Commune")

#calculating the data for the 0-19 years residents with appropriate percentages, so it returns true if the conditions are met.
Fivetoten_percent = (df["0-19 years"] > 5) & (df["0-19 years"] < 10)
Tentofifteen_percent = (df["0-19 years"] > 10) & (df["0-19 years"] < 15)
Fifteentotwenty_percent = (df["0-19 years"] > 15) & (df["0-19 years"] < 20)
Twentytotwentyfive_percent = (df["0-19 years"] > 20) & (df["0-19 years"] < 25)
Twentyfivetothirty_percent = (df["0-19 years"] > 25) & (df["0-19 years"] < 30)
Thirtytothirtyfive_percent = (df["0-19 years"] > 30) & (df["0-19 years"] < 35)

#Setting the width and height for the figure.
plt.figure(figsize=(5, 4))
#Getting the respective values from the data frame for the variable when the conditions are met and plotting the histogram.
plt.hist(
    [df.loc[Fivetoten_percent, "0-19 years"], df.loc[Tentofifteen_percent, "0-19 years"],
     df.loc[Fifteentotwenty_percent, "0-19 years"], df.loc[Twentytotwentyfive_percent, "0-19 years"],
     df.loc[Twentyfivetothirty_percent, "0-19 years"],df.loc[Thirtytothirtyfive_percent, "0-19 years"]],
    stacked=True,
    edgecolor="black",
    #bins=20
    bins=[5,10,15,20,25,30,35],   
)
         
plt.legend(["5%-10%", "10%-15%", "15%-20%", "20%-25%", "25%-30%", "30%-35%"], loc="upper right")
plt.xlabel("0-19 years residents")
plt.ylabel("Number of Commune")
plt.title("Distribution of '0-19 years' young residents")
plt.show()
```

# E. Matrices : 

#### Define a data frame matrix whose rows correspond to communes and the columns to the cantons. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the commune in row i is in the canton in column j and a 0 otherwise.

Just as an example got the first 100 values of the commune and canton from the csv file and according to their shapes defined a matrix with zeroes. Then looped through the two arrays to set the corresponding value of the matrix with 1 or 0 if the value of the commune is contained in the canton or the other way round.



```python
df = pd.read_csv("p1_communes.csv")
df.dropna(inplace = True) #Skipped the null values
rows = df["Commune"].head(100).to_numpy() 
columns = df["Canton"].head(100).to_numpy() 

#Tried with this example:
#rows = ["Epalinges", "Bussigny", "Lausanne", "Prilly", "Biel-VS" ]
#columns = ["VD", "VD", "VD", "VD", "VS"]

#Defining a matrix with the given shape  
matrix =  np.zeros((len(rows),len(columns)))

#Looping the arrays of Commune and Canton and filling the matrix with 1 if the value of commune is contained in Canton or the other way round.
for j in range(len(columns)):
    for i in range(len(rows)):
        if rows[i] in columns[j] or columns[j] in rows[i]:
            matrix[i,j]=1
        else:
            matrix[i, j]=0
        
print(matrix)
```
